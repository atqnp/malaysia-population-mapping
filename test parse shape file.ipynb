```python
#Pandas Dataframe of the shp file
#the column is rename for merging purposes
import geopandas
import pandas as pd

df = geopandas.read_file("gadm36_MYS_2.shp")
df = df[['NAME_1','NAME_2','HASC_2','geometry']]
df= df.rename(index=str, columns ={'NAME_1':'State','NAME_2':'District','HASC_2':'HASC','geometry':'geo'})
df
```

```python
#parsing url to obtain the population data
import requests
import pandas as pd
from bs4 import BeautifulSoup
url = requests.get('http://www.statoids.com/ymy.html')

dist_list=[]
header = []
soup = BeautifulSoup(url.content,'lxml')
table_class = {'class': ['st']}
table = soup.find('table',table_class)
rows = table.findAll('tr')
#content
for row in rows:
    cols=row.find_all('td')
    cols=[x.text.strip() for x in cols]
    dist_list.append(cols)
#header
for row in rows:
    cols=row.find_all('th')
    cols=[x.text.strip() for x in cols]
    header.append(cols)
header[0]

df_pop = pd.DataFrame(dist_list, columns=header[0])
df_pop=df_pop.drop(145)
df_pop=df_pop.dropna(how='all')
df_pop
```

```python
#merge both Dataframe to one
result = pd.merge(df,df_pop, on = ['HASC'],how='outer')
result = result[['State','District_x','District_y','HASC','Pop-2010','Pop-2000','Pop-1991','Division','geo']]
result
```

```python
#exporting the result as csv
result.to_csv("malaysia_area.csv",index=None)
```
